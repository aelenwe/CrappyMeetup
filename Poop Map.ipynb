```python
from __future__ import division
import counts
import folium  
from folium import features, plugins
import time
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
from my_map import draw_map
from util import radius, months, days
import mycredentials
user, key = mycredentials.plotly_cred()
import plotly.plotly as py
py.sign_in(user,key)
import plotly.graph_objs as go
```

```python
year = [int(x[0]) for x in counts.counts_by_year()]
count = [int(x[1]) for x in counts.counts_by_year()]
data = [go.Bar(x=year,y=count)]
layout = go.Layout(title='Reports of Human Waste By Year',
                   xaxis=dict(title='Year', titlefont=dict(family='Open Sans', size=18, color='#7f7f7f')),
                   yaxis=dict(title='Number of Reports', titlefont=dict(family='Open Sans', size=18, color='#7f7f7f')))
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='basic-bar')
```

```python
# Create a slide bar for how many months of 311 history to include
months = interactive(months, months=widgets.IntSlider(min=1,max=96,step=1,value=6))
# Create a slide bar for radius around meetup location to look for human waste
radius = interactive(radius, radius=widgets.IntSlider(min=1,max=500,step=1,value=100))
# Create a slide bar for number of days of Meetups to include
days = interactive(days, days=widgets.IntSlider(min=1,max=500,step=1,value=14))
```
